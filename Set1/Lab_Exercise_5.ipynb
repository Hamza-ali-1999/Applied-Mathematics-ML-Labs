```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

The process below shows how to implement a simple linear regression analysis for a vehicle pricing scenario. 

Task 1: Re-run it and take note of the R-squared value. 
Task 2: Then, change the std value in np.random.normal to 1, 5 and 10. Observe the change in R-squared. 
Question: How do you explain the change in R-squared value?

```python
vehicle_age = np.random.uniform(0,10,1000)
vehicle_price = 70000 - (vehicle_age + np.random.normal(0, 1, 1000)) * 3000

vehicle_age = vehicle_age.reshape(-1, 1)  # Reshape for single feature, use (1,-1) for single sample
vehicle_price = vehicle_price.reshape(-1, 1)

plt.scatter(vehicle_age, vehicle_price)
```

```python
reg = LinearRegression().fit(vehicle_age, vehicle_price)
reg.score(vehicle_age, vehicle_price)
```
